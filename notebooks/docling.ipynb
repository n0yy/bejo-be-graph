```python
from docling.document_converter import DocumentConverter
```

# DOCX

```python
source = "documents_1.docx"
converter = DocumentConverter()
result_docx = converter.convert(source)

markdown_result = result_docx.document.export_to_markdown()
print(markdown_result)
```

# PPTX

```python
result_pptx = converter.convert("document_2.pptx")
markdown_result = result_pptx.document.export_to_markdown()
print(markdown_result)
```

# XLSX

```python
result_xlsx = converter.convert("document_3.xlsx")
markdown_result = result_xlsx.document.export_to_markdown()

print(markdown_result)
```

# CSV

```python
result_csv = converter.convert("document_4.csv").document

print("Schema Name: ", result_csv.schema_name)
print("Version: ", result_csv.version)
print("Name: ", result_csv.name)

# Metadata asal dockument
print("MIME Type: ", result_csv.origin.mimetype)
print("Binary Hash: ", result_csv.origin.binary_hash)
print("Filename: ", result_csv.origin.filename)
print("URI: ", result_csv.origin.uri)
```

# Convert into Document Langchain

```python
from langchain_core.documents import Document
```

```python
source = "documents_1.docx"
result = converter.convert(source).document

document = Document(
    page_content=result.export_to_markdown(),
    metadata={"source": result.origin.uri, "filename": result.origin.filename, "mimetype": result.origin.mimetype},
)

document
```

# Chunking

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len
)

docs_splitted = splitter.split_documents([document])
docs_splitted[2]
```
